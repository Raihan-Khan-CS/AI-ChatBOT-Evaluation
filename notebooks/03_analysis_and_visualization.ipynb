```python
import sys
sys.path.insert(0, '..')

import yaml
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100

from src.statistical_analysis import StatisticalAnalyzer
from src.visualizer import Visualizer
```

```python
with open('../configs/config.yaml', 'r') as f:
    config = yaml.safe_load(f)

# Load scores
scores_df = pd.read_csv('../data/results/evaluation_scores.csv')
print(f'Loaded {len(scores_df)} evaluation scores')
print(f'Chatbots: {scores_df["chatbot_id"].unique()}')
scores_df.head()
```

## 1. Descriptive Statistics

```python
# Overall performance summary
summary = scores_df.groupby('chatbot_id')[[
    'accuracy_raw', 'authenticity_raw', 'uptodate_raw', 'overall_raw'
]].describe().round(4)
summary
```

```python
# Accuracy and Up-to-dateness rates
rates = scores_df.groupby('chatbot_id').agg(
    accuracy_rate=('is_accurate', 'mean'),
    authenticity_rate=('is_authentic', 'mean'),
    uptodate_rate=('is_uptodate', 'mean'),
    count=('query_id', 'count')
).round(4)
print('Performance Rates:')
rates
```

## 2. Statistical Analysis

```python
analyzer = StatisticalAnalyzer(config)
stat_results = analyzer.run_full_analysis(scores_df)
analyzer.print_summary(stat_results)
```

## 3. Visualizations

```python
viz = Visualizer(config)
viz.generate_all(scores_df, stat_results)
print('All figures generated!')
```

```python
# Display generated figures inline
import os
from IPython.display import Image, display

figures_dir = '../data/results/figures/'
for fig_file in sorted(os.listdir(figures_dir)):
    if fig_file.endswith('.png'):
        print(f'\n--- {fig_file} ---')
        display(Image(filename=os.path.join(figures_dir, fig_file), width=800))
```
