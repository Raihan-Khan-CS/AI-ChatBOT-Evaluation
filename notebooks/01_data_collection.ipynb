## 1. Explore the Query Dataset

```python
qm = QueryManager('../data/queries/evaluation_queries.json')
queries = qm.get_all_queries()

print(f'Total queries: {len(queries)}')
print(f'\nStatistics:')
stats = qm.get_statistics()
for key, value in stats.items():
    print(f'  {key}: {value}')
```

```python
# View queries as DataFrame
queries_df = pd.DataFrame(queries)
queries_df[['id', 'domain', 'type', 'difficulty', 'question']].head(15)
```
